```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Cargamos dataset Titanic desde Seaborn
df = sns.load_dataset('titanic')
df.head()
```

```
# Cantidad de datos
df.shape
```

```
#Cantidad de datos nulos
df.isnull().sum()
```

¿Afectara el sexo a la supervivencia?

```
#primero graficamos cantidades de personas separadas por sexo
plt.subplots(figsize=(7,5))
df['sex'].value_counts().plot.bar(color = ['b','r'])
plt.show()
```

```
#tasa de supervivencia femenima
female_survived=df.loc[df['sex']== 'female']['survived']
rate_female = sum(female_survived)/len(female_survived)
rate_female
```

```
#tasa de supervivencia masculina
male_survived=df.loc[df['sex']== 'male']['survived']
rate_male = sum(male_survived)/len(male_survived)
rate_male
```

```
#Convertir categoria numerica survived en categorica 
df['name_survived']= df['survived'].map({0:'no_survived',1:'survived'})
df
```

```
df['name_survived'].astype('category')
df
```
